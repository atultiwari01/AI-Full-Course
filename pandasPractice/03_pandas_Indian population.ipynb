 ## **Dataset Overview**  
This dataset provides population statistics of Indian states and union territories (UTs) from 1951 to 2011, based on census data. It includes population figures for each decade, making it useful for analyzing population growth trends across different regions of India.  

## **Columns Description**  
- **Rank**: Rank of the state/UT based on the population in 2011.  
- **State or Union Territory**: Name of the state or UT.  
- **Population 1951**: Population recorded in the 1951 census.  
- **Population 1961 - Population 2011**: Population data for each decade from 1961 to 2011.  

## **Potential Use Cases**  
- Analyzing demographic trends in India.  
- Forecasting future population growth.  
- Regional policy-making and resource allocation.  
- Studying urbanization and migration patterns.  

## **Source**  
This dataset is derived from official Indian census reports.

**Working Population of India**

```python
total_pop = df['Population'].sum()
total_Working_pop = (df['Male_Workers'].sum())+(df['Female_Workers'].sum())
print(f'Total Population :{total_pop}')
print(f'Total working Population :{total_Working_pop}')
print(f'Total Unemployed Population :{(total_pop)-(total_Working_pop)}')
```

**Population Distribution based on Religion**

```python
# Total population of 'Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains'

print(f"Population of Hindus : {df['Hindus'].sum()}")
print(f"Population of Muslims : {df['Muslims'].sum()}")
print(f"Population of Christians : {df['Christians'].sum()}")
print(f"Population of Sikhs : {df['Sikhs'].sum()}")
print(f"Population of Buddhists : {df['Buddhists'].sum()}")
print(f"Population of Jains : {df['Jains'].sum()}")
```

```python
# Percentage Distribution
print(f"Population of Hindus : {(df['Hindus'].sum())/(total_pop)*100}")
print(f"Population of Muslims : {df['Muslims'].sum()/(total_pop)*100}")
print(f"Population of Christians : {df['Christians'].sum()/(total_pop)*100}")
print(f"Population of Sikhs : {df['Sikhs'].sum()/(total_pop)*100}")
print(f"Population of Buddhists : {df['Buddhists'].sum()/(total_pop)*100}")
print(f"Population of Jains : {df['Jains'].sum()/(total_pop)*100}")
```
